# **QUESTION 1 --  What is Simple Linear Regression?**

# **ANSWER 1 :** Simple Linear Regression is a statistical method that allows us to model the relationship between two variables: a dependent variable (also known as the response or outcome variable) and an independent variable (also known as the predictor or explanatory variable). It assumes a linear relationship, meaning that the change in the dependent variable is directly proportional to the change in the independent variable.

 The Core Idea: TheThe fundamental idea is to find a straight line that best fits the data points. This line is represented by the equation:

Y = β₀ + β₁X + ε

Where:

Y is the dependent variable.

X is the independent variable.

β₀ (beta-naught) is the Y-intercept. It represents the predicted value of Y when X is 0.

β₁ (beta-one) is the slope of the line. It represents the change in Y for a one-unit change in X.

ε (epsilon) is the error term or residual. It accounts for the variability in Y that cannot be explained by X.

How it Works (Least Squares Method): The goal is to find the values of β₀ and β₁ that minimize the sum of the squared differences between the observed Y values and the predicted Y values (from the regression line). This method is called the Ordinary Least Squares (OLS) method. By minimizing these squared errors, we ensure that the line is as close as possible to all data points.

Assumptions of Simple Linear Regression: For the results of a simple linear regression model to be reliable and interpretable, several assumptions should ideally be met:

Linearity: The relationship between the independent and dependent variables should be linear. If not, the model may not accurately capture the relationship.
Independence of Errors: The errors (residuals) should be independent of each other. This means that the residual for one observation does not affect the residual for another.

Homoscedasticity: The variance of the errors should be constant across all levels of the independent variable. In simpler terms, the spread of the residuals should be roughly the same for all predicted values.
Normality of Errors: The errors (residuals) should be normally distributed. This assumption is particularly important for hypothesis testing and confidence interval estimation.

No Outliers: The data should not contain significant outliers that can heavily influence the regression line.
Applications: Simple Linear Regression is widely used for:

Prediction: Forecasting future values of the dependent variable based on the independent variable.

Relationship Analysis: Understanding the strength and direction of the relationship between two variables.


Trend Analysis: Identifying trends over time or across different conditions.
For example, you might use simple linear regression to predict a student's test score based on the number of hours they studied, or to see how the price of a house changes with its size. While 'simple,' it forms the bedrock for more complex regression models.





#**Question 2: What are the key assumptions of Simple Linear Regression?**

#**ANSWER 2:**
# ASSUMPTIONS OF SIMPLE LINEAR REGRESSION

#**Linearity:**
 This is the most fundamental assumption. It states that the relationship between the independent variable (X) and the dependent variable (Y) is linear. If you were to plot your data points, they should roughly form a straight line, not a curve or some other complex shape.

 If the true relationship is non-linear, a linear model will fail to capture the actual pattern, leading to biased estimates and poor predictions. You might misinterpret the nature and strength of the relationship

  Visual inspection of scatter plots of Y versus X. Residual plots (residuals vs. predicted values) can also help; if there's a pattern (e.g., a curve) in the residuals, linearity might be violated.

#**Independence of Errors (or Residuals):**

This assumption implies that the error term (ε) for one observation is not correlated with the error term of any other observation. In simpler terms, the prediction error for one data point should not provide any information about the prediction error for another data point.

Violations often occur in time series data (where errors from one period might affect the next) or clustered data. Independent errors are crucial for accurate standard errors, p-values, and confidence intervals. If errors are correlated, your model might appear more precise than it actually is.

 For time series data, a Durbin-Watson test can be used. Residual plots over time can also reveal patterns.

#**Homoscedasticity:**

 This assumption states that the variance of the error terms (residuals) is constant across all levels of the independent variable. In other words, the spread of the residuals should be roughly the same for small values of X, medium values of X, and large values of X.

 If homoscedasticity is violated (a condition called heteroscedasticity), the OLS estimators are still unbiased, but they are no longer the most efficient. This leads to incorrect standard errors, making hypothesis tests and confidence intervals unreliable. The model might give more weight to observations from regions with larger error variance.

 Visual inspection of residual plots (residuals vs. predicted values or residuals vs. independent variable). You should see a random scatter of points without any funnel or cone shape.

#**Normality of Errors (or Residuals):**

This assumption requires that the error terms are normally distributed. It doesn't mean that X or Y themselves have to be normally distributed, only the errors around the regression line.

 While OLS estimates of the regression coefficients (β₀ and β₁) are robust to departures from normality for large sample sizes (due to the Central Limit Theorem), this assumption is critical for the validity of hypothesis tests (t-tests, F-tests) and the construction of confidence intervals. If errors are not normal, especially with small sample sizes, inferences about the population parameters might be incorrect.

Histograms of residuals, Q-Q plots (quantile-quantile plots) of residuals, or statistical tests like the Shapiro-Wilk test or Kolmogorov-Smirnov test.

#**No Outliers:**

 The data should not contain significant outliers or influential points that can disproportionately affect the regression line.

 Outliers can pull the regression line significantly towards themselves, leading to biased coefficient estimates and inflated standard errors, thus misrepresenting the true relationship between the variables for the majority of the data. Influential points are observations that, if removed, would significantly change the coefficients of the regression model.

 Scatter plots, residual plots (looking for points far from the main cloud), Cook's distance, leverage values, and DFFITS statistic.
Understanding and checking these assumptions are crucial steps in building a reliable and interpretable simple linear regression model

#**Question 3: What is heteroscedasticity, and why is it important to address in regression models?**

#**ANSWER 3 -**
Heteroscedasticity is a specific violation of one of the key assumptions of Ordinary Least Squares (OLS) regression: homoscedasticity. Homoscedasticity assumes that the variance of the error term (residuals) is constant across all levels of the independent variables. When this assumption is violated, we have heteroscedasticity, meaning the spread of the residuals is not constant.

Imagine plotting your residuals against the predicted values (or against one of your independent variables). In a homoscedastic scenario, the points would form a random, consistent band around zero. With heteroscedasticity, you'd typically see a pattern, such as:

Fanning out: The spread of residuals increases as the predicted value (or independent variable) increases.

Fanning in: The spread of residuals decreases as the predicted value (or independent variable) increases.

Other systematic patterns: Like a bow-tie shape or clusters
.
Why is it so important to address?

Inefficient Estimators: While OLS estimators (your β₀ and β₁ coefficients) remain unbiased in the presence of heteroscedasticity, they are no longer the 'Best Linear Unbiased Estimators' (BLUE). This means there are other linear unbiased estimators that could estimate the true population parameters with smaller variance. In practical terms, your OLS estimates are not as precise as they could be.


Incorrect Standard Errors: This is the most critical consequence. The formulas used to calculate the standard errors of your regression coefficients assume homoscedasticity. When heteroscedasticity is present, these standard errors will be biased:

They are often underestimated when there's positive correlation between the squared errors and the predictors. This makes your t-statistics larger, leading to a higher chance of incorrectly rejecting the null hypothesis (Type I error). You might conclude a variable is statistically significant when it isn't.
They can be overestimated in other cases, making your t-statistics smaller, and potentially leading to a higher chance of incorrectly failing to reject the null hypothesis (Type II error). You might miss a truly significant relationship.

Invalid Hypothesis Tests and Confidence Intervals: Because standard errors are incorrect, any hypothesis tests (like t-tests for individual coefficients or F-tests for overall model significance) become invalid. Similarly, confidence intervals for the coefficients will be too narrow or too wide, meaning they do not accurately reflect the true uncertainty around your estimates. This severely compromises your ability to make reliable statistical inferences about the population.


Misleading p-values: The p-values derived from these incorrect standard errors will also be wrong, further undermining the validity of your conclusions about statistical significance.

Biased Model Selection: If you are comparing multiple models using criteria that rely on correctly estimated standard errors (e.g., AIC, BIC), heteroscedasticity can lead to incorrect model selection.

Common Causes of Heteroscedasticity:

Measurement Error: Errors in measuring the dependent variable might increase with the magnitude of the independent variable.

Learning/Improvement: In studies of learning, errors might be larger at the beginning and decrease over time as subjects improve.

Data Aggregation: Aggregated data (e.g., across firms of different sizes) often exhibits heteroscedasticity.

Incorrect Functional Form: If the linear model is not the correct functional form for the relationship, heteroscedasticity can appear in the residuals.

Presence of Outliers: Outliers can disproportionately affect the variance of residuals.



#**QUESTION 4: What is Multiple Linear Regression?**

#**ANSWER 4:**
Multiple Linear Regression is an extension of Simple Linear Regression that allows for the modeling of a linear relationship between a dependent variable and two or more independent variables. While simple linear regression examines the relationship between one dependent variable and one independent variable, multiple linear regression enables us to consider the influence of several independent variables simultaneously on the dependent variable.

**The Core Idea:** Instead of fitting a straight line in a 2D space, multiple linear regression fits a hyperplane (a plane in higher dimensions) to the data points. The goal remains to find the best-fitting linear combination of independent variables that predicts the dependent variable.

**The Equation:** The general form of the multiple linear regression equation is:

Y = β₀ + β₁X₁ + β₂X₂ + ... + βₚXₚ + ε

Where:

Y is the dependent variable (the variable you are trying to predict).
X₁, X₂, ..., Xₚ are the p independent variables (the predictors).\

β₀ is the Y-intercept. It represents the predicted value of Y when all independent variables (X values) are zero.

β₁, β₂, ..., βₚ are the coefficients for each independent variable. Each βᵢ represents the change in Y for a one-unit change in Xᵢ, while holding all other independent variables constant.

ε (epsilon) is the error term or residual, representing the unobserved factors that influence Y and the random variability.

#Key Differences from Simple Linear Regression:

**Number of Independent Variables:** Simple Linear Regression uses only one independent variable, whereas Multiple Linear Regression uses two or more.

**Complexity of Relationship:** Multiple Linear Regression can model more complex relationships by accounting for the combined and individual effects of multiple predictors.

**Interpretation of Coefficients:** In simple linear regression, β₁ is the direct change in Y per unit change in X. In multiple linear regression, each βᵢ is a partial regression coefficient, indicating the change in Y for a one-unit change in Xᵢ while controlling for the effects of all other independent variables in the model.

#Purpose and Applications:

**Prediction:** To forecast the value of the dependent variable based on the values of multiple independent variables (e.g., predicting house prices based on size, number of bedrooms, and location).

**Explanation:** To understand the individual and combined influence of different independent variables on the dependent variable (e.g., identifying factors that contribute to student performance).

**Control for Confounding Factors:** By including multiple variables, you can statistically control for the influence of other factors when examining the relationship between a specific independent variable and the dependent variable.

**Assumptions:** Multiple Linear Regression shares most of the same assumptions as Simple Linear Regression, extended to multiple independent variables:

**Linearity:** A linear relationship between the dependent variable and each independent variable, as well as their combined effect.

**Independence of Errors:** Errors are independent of each other.
Homoscedasticity: Constant variance of errors across all levels of independent variables.

**Normality of Errors:** Errors are normally distributed.
No Multicollinearity: Independent variables should not be too highly correlated with each other (this is a new, crucial assumption for multiple regression).

**No Outliers:** Absence of significant outliers or influential points.
Multiple linear regression is a powerful and widely used statistical tool for analyzing complex relationships in various fields like economics, social sciences, engineering, and healthcare.




#**Question 5: What is polynomial regression, and how does it differ from linear regression**

#**ANSWER 5 :**
Polynomial Regression is a form of regression analysis in which the relationship between the independent variable (X) and the dependent variable (Y) is modeled as an nth-degree polynomial. While it still falls under the umbrella of linear models (because it is linear in the parameters/coefficients), it can capture non-linear relationships between the variables.

The Equation: The general form of a polynomial regression equation is:

Y = β₀ + β₁X + β₂X² + ... + βₚXᵖ + ε

Where:

Y is the dependent variable.

X is the independent variable.

β₀, β₁, ..., βₚ are the coefficients (parameters) of the polynomial.
p is the degree of the polynomial, which determines the curvature of the line.
ε is the error term.

For example, if p=2, it's a quadratic relationship (Y = β₀ + β₁X + β₂X² + ε), which results in a parabolic curve.

**How does it differ from Linear Regression?**

The fundamental difference lies in the form of the relationship they can model between the independent and dependent variables, even though both are 'linear' in terms of their coefficients.

Form of the Relationship:

Linear Regression: Assumes a straight-line relationship between the independent and dependent variables. The model is Y = β₀ + β₁X + ε.

Polynomial Regression: Models a curvilinear relationship by introducing polynomial terms (X², X³, etc.). This allows the model to fit curves rather than just straight lines.

#Flexibility:

Linear Regression: Less flexible, suitable only when the relationship is genuinely linear.

Polynomial Regression: More flexible, capable of fitting a wider range of shapes and patterns in the data, making it useful when the relationship is non-linear.

#Interpretation:

#**Linear Regression:**
 The coefficient β₁ directly represents the constant change in Y for a one-unit change in X. Interpretation is straightforward.

**Polynomial Regression:**
The interpretation of individual coefficients (β₁, β₂, etc.) becomes less direct. For instance, β₁ no longer represents a constant change in Y per unit change in X; rather, the effect of X on Y depends on the current value of X (due to the X² or higher-order terms). A higher degree polynomial can also make interpretation more complex due to multicollinearity between the polynomial terms (e.g., X and X² are often correlated).

#**Overfitting Risk:**

**Linear Regression:**
 Generally less prone to overfitting if the true relationship is linear, but underfits if it's non-linear.

**Polynomial Regression:** Has a higher risk of overfitting, especially with higher degrees of polynomials. A very high-degree polynomial can fit the training data almost perfectly but perform very poorly on new, unseen data, as it models the noise in the training data rather than the underlying pattern.

**When to use Polynomial Regression:** You would typically consider polynomial regression when you observe a non-linear pattern in your scatter plot of the dependent and independent variables, and you believe that a simple curved relationship (like a parabola or cubic curve) can adequately capture that pattern. It's a way to extend the power of linear models to fit more complex data distributions without resorting to completely non-linear modeling techniques.

In essence, while linear regression aims to fit the 'best straight line' through the data, polynomial regression aims to fit the 'best curve' through the data, defined by a polynomial equation.




#**Question 6: Implement a Python program to fit a Simple Linear Regression model to**

**the following sample data:**

**● X = [1, 2, 3, 4, 5]**

**● Y = [2.1, 4.3, 6.1, 7.9, 10.2]**

**Plot the regression line over the data points.**


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5])
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Reshape X to be a 2D array, as scikit-learn expects
X_reshaped = X.reshape(-1, 1)

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X_reshaped, Y)

# Get the predicted Y values
Y_pred = model.predict(X_reshaped)

# Get the coefficients (slope and intercept)
slope = model.coef_[0]
intercept = model.intercept_

print(f"Intercept (β0): {intercept:.2f}")
print(f"Slope (β1): {slope:.2f}")

# Plot the data points and the regression line
plt.figure(figsize=(8, 6))
plt.scatter(X, Y, color='blue', label='Actual Data Points')
plt.plot(X, Y_pred, color='red', label=f'Regression Line (Y = {intercept:.2f} + {slope:.2f}X)')
plt.title('Simple Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```

#**Question 7: Fit a Multiple Linear Regression model on this sample data:**

**● Area = [1200, 1500, 1800, 2000]**

**● Rooms = [2, 3, 3, 4]**

**● Price = [250000, 300000, 320000, 370000]**

**Check for multicollinearity using VIF and report the results**.
**(Include your Python code and output in the code box below.)**

#**ANSWER 7:**

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm

# Sample data
Area = np.array([1200, 1500, 1800, 2000])
Rooms = np.array([2, 3, 3, 4])
Price = np.array([250000, 300000, 320000, 370000])

# Create a DataFrame for the independent variables (features)
X_df = pd.DataFrame({'Area': Area, 'Rooms': Rooms})
y = Price

# --- Fit Multiple Linear Regression Model ---
# Initialize and fit the model
model_mlr = LinearRegression()
model_mlr.fit(X_df, y)

# Get coefficients and intercept
intercept_mlr = model_mlr.intercept_
coefficients_mlr = model_mlr.coef_

print("\n--- Multiple Linear Regression Results ---")
print(f"Intercept: {intercept_mlr:.2f}")
print(f"Coefficient for Area: {coefficients_mlr[0]:.2f}")
print(f"Coefficient for Rooms: {coefficients_mlr[1]:.2f}")

# --- Check for Multicollinearity using VIF ---
# Add a constant term to the independent variables for VIF calculation
X_vif = sm.add_constant(X_df)

# Calculate VIF for each independent variable
vif_data = pd.DataFrame()
vif_data["feature"] = X_vif.columns
vif_data["VIF"] = [variance_inflation_factor(X_vif.values, i) for i in range(X_vif.shape[1])]

print("\n--- Multicollinearity Check (VIF) ---")
print(vif_data)

# Interpret VIF results
print("\n--- VIF Interpretation ---")
print("A common rule of thumb is:")
print("- VIF = 1: No correlation between the feature and other features.")
print("- 1 < VIF < 5: Moderate correlation, generally considered acceptable.")
print("- VIF >= 5 (or 10): High correlation, indicating potential multicollinearity issues.")
print("Based on the VIF values above, you can assess the degree of multicollinearity in the model.")
```

#**Question 8: Implement polynomial regression on the following data:**

**● X = [1, 2, 3, 4, 5]**


**● Y = [2.2, 4.8, 7.5, 11.2, 14.7]**

**Fit a 2nd-degree polynomial and plot the resulting curve.**

#**ANSWER 8 :**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Sample data
X = np.array([1, 2, 3, 4, 5])
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Reshape X to be a 2D array, as scikit-learn expects
X_reshaped = X.reshape(-1, 1)

# Create polynomial features for a 2nd-degree polynomial
# include_bias=False ensures that we don't duplicate the intercept term
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X_reshaped)

# Create a linear regression model
model_poly = LinearRegression()

# Fit the model to the polynomial features
model_poly.fit(X_poly, Y)

# Get the coefficients and intercept
intercept_poly = model_poly.intercept_
coefficients_poly = model_poly.coef_

print("--- Polynomial Regression Results (Degree 2) ---")
print(f"Intercept (β0): {intercept_poly:.2f}")
print(f"Coefficients (β1, β2): {coefficients_poly}")

# Generate points for the smooth curve
X_plot = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
X_plot_poly = poly_features.transform(X_plot)
Y_plot_pred = model_poly.predict(X_plot_poly)

# Plot the actual data points and the polynomial regression curve
plt.figure(figsize=(8, 6))
plt.scatter(X, Y, color='blue', label='Actual Data Points')
plt.plot(X_plot, Y_plot_pred, color='red', label='Polynomial Regression Curve (Degree 2)')
plt.title('Polynomial Regression (Degree 2)')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```

#**Question 9: Create a residuals plot for a regression model trained on this data:**

**● X = [10, 20, 30, 40, 50]**

**● Y = [15, 35, 40, 50, 65]**

**Assess heteroscedasticity by examining the spread of residuals.**

**(Include your Python code and output in the code box below.)**


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([10, 20, 30, 40, 50])
Y = np.array([15, 35, 40, 50, 65])

# Reshape X to be a 2D array, as scikit-learn expects
X_reshaped = X.reshape(-1, 1)

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X_reshaped, Y)

# Get the predicted Y values
Y_pred = model.predict(X_reshaped)

# Calculate residuals
residuals = Y - Y_pred

# Plot the residuals
plt.figure(figsize=(10, 6))
plt.scatter(X, residuals, color='blue') # Plot residuals against X
plt.axhline(y=0, color='red', linestyle='--') # Add a horizontal line at y=0
plt.title('Residuals Plot')
plt.xlabel('Independent Variable (X)')
plt.ylabel('Residuals')
plt.grid(True)

# Annotate for heteroscedasticity assessment
plt.text(X.min(), residuals.max() * 1.1,
         "Assess heteroscedasticity by observing the spread of residuals.\n" \
         "A random, consistent band around zero indicates homoscedasticity.\n" \
         "A fanning out/in or other systematic pattern indicates heteroscedasticity.",
         horizontalalignment='left', verticalalignment='top', fontsize=9, bbox=dict(boxstyle="round,pad=0.3", fc="yellow", ec="b", lw=1, alpha=0.5))

plt.show()

print("\n--- Heteroscedasticity Assessment ---")
print("Examine the residuals plot above. If the points are randomly scattered around the zero line with a roughly constant width, the assumption of homoscedasticity holds.")
print("If the spread of the residuals increases or decreases as X increases (e.g., a funnel or cone shape), then heteroscedasticity is present.")
```
